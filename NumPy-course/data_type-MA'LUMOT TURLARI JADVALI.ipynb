# Massivlar yaratishda ma'lumotlar turini aniq kiritish

```python
import numpy as np
```

```python
arr1 = np.array([1, 3, 5], dtype = np.float64) # bu float64 turiga ega massiv
arr1
```

```python
arr2 = np.array([1, 3, 5], np.int32) # int32 malumotlar turiga ega massiv
arr2
```

# Malumotlar turini aliqlash 
 - .dtype

```python
arr1.dtype
```

```python
arr2.dtype
```

# Bir turdan ikkinchi turga o'girish
 - .astype

```python
float_data = arr2.astype(np.float64) # yuqoridagi int turdagi ma'lumotni float turiga o'qirish
float_data.dtype
```

```python
int_data = arr1.astype(np.int64) # float turini intga o'girish
int_data.dtype
```
